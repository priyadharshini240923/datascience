```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 20, 10)
y = np.sin(x)

plt.plot(x, y, label = "sine wave")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("simple line plot")
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
vales = [10,25,12,30]
plt.figure(figsize=(4,4))
plt.bar(categories,vales,color='purple')
plt.xlabel("categories")
plt.ylabel("values")
plt.title("Bar Chart Example")
plt.show()
```

```python
data = np.random.randn(1000)
plt.figure(figsize=(7,5))
plt.hist(data,bins=30,color='green',edgecolor='black',alpha=0.7)
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("histogram example")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Creating sample data
data = np.random.randn(1000)
df = pd.DataFrame(data, columns=['Values'])

# Plot
sns.histplot(df['Values'], bins=30, kde=True, color='blue')
plt.title("Histogram with KDE")
plt.show()
```

```python
tips = sns.load_dataset('tips')

plt.figure(figsize=(6, 4))
sns.boxplot(x=tips['total_bill'])
plt.title("Box Plot of Total Bill")
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
corr_matrix=tips.select_dtypes(include=np.number).corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt=".2f")
plt.title("correlation heatmap")
plt.show()
```

```python
import plotly.express as px

df = pd.DataFrame({
    "x": np.linspace(0, 10, 100),
    "y": np.sin(np.linspace(0, 10, 100))
})

fig = px.line(df, x='x', y='y', title="Interactive Sine Wave")
fig.show()
```

```python
 ig = px.scatter(tips, x='total_bill', y='tip', color='sex', size='size',title="Total Bill vs Tip")
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter3d(
    x=tips['total_bill'],
    y=tips['tip'],
    z=tips['size'],
    mode='markers',
    marker=dict(size=5, color=tips['total_bill'], colorscale='Viridis')
)])

fig.update_layout(title="3D Scatter Plot of Total Bill, Tip & Size")
fig.show()
```
